```python
def secant_method(f, x0, x1, tol=1e-6, max_iter=100):    
    x_prev = x0
    x_curr = x1
    iter_count = 0
    f_prev = f(x_prev)
    f_curr = f(x_curr)
    
    while abs(f(x_curr)) > tol and iter_count < max_iter:
        denominador = f_curr - f_prev
        if denominador == 0:
            raise ValueError("Denominator in secant method became zero.")
        # Calculate the next approximation using the secant method formula
        x_next = x_curr - (f_prev * (x_curr - x_prev)) / denominador

        # Update variables for the next iteration
        x_prev, f_prev = x_curr, f_curr
        x_curr = x_next
        f_curr = f(x_next)
        iter_count += 1

    return x_curr, iter_count
```
